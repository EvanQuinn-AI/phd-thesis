```python
import imaplib
import email
import mysql.connector
from datetime import datetime
import time
import os
import re
import os.path
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import base64
from email import message_from_bytes

# If modifying these scopes, delete the file token.json.
SCOPES = ["https://www.googleapis.com/auth/gmail.readonly"]
EMAILS = None

def gmail_setup():
    """Shows basic usage of the Gmail API.
    Lists the user's Gmail labels.
    """
    creds = None
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists("token.json"):
        creds = Credentials.from_authorized_user_file("token.json", SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                "credentials.json", SCOPES
            )
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open("token.json", "w") as token:
            token.write(creds.to_json())

    try:
        # Call the Gmail API
        service = build("gmail", "v1", credentials=creds)
        # Fetch the list of labels in the user's account
        results = service.users().labels().list(userId="me").execute()
        labels = results.get("labels", [])

        # read emails
        # Now, let's list all emails in the inbox
        print("\nFetching emails from the inbox...\n")

        # Fetch messages in the inbox
        result = service.users().messages().list(userId='me', labelIds=['INBOX'], maxResults=10).execute()
        messages = result.get('messages', [])

        if not messages:
            print("No messages found.")
            return

        for msg in messages:
            msg_id = msg['id']  # Get the message ID
            message = service.users().messages().get(userId='me', id=msg_id, format='raw').execute()
            
            # Decode the message from base64
            msg_str = base64.urlsafe_b64decode(message['raw'].encode('ASCII'))
            mime_msg = message_from_bytes(msg_str)

            EMAILS.append(message)
            
            # Extract the email details
            subject = mime_msg['subject']
            sender = mime_msg['from']
            print(f"From: {sender}")
            print(f"Subject: {subject}")
            
            # If the email is multipart, print the text/plain content
            if mime_msg.is_multipart():
                for part in mime_msg.walk():
                    if part.get_content_type() == 'text/plain':
                        body = part.get_payload(decode=True).decode()
                        print(f"Body: {body}")
            else:
                # If it's not multipart, just decode the message payload
                body = mime_msg.get_payload(decode=True).decode()
                print(f"Body: {body}")
            
            print("-" * 40)

    except HttpError as error:
        # Handle errors from the Gmail API
        print(f"An error occurred: {error}")

# Function to connect to MySQL database
def connect_to_mysql():
    try:
        conn = mysql.connector.connect(
            host='192.168.6.5',
            port=3306,
            user='root',
            password='',
            database='stratocal'
        )
        print("Connected successfully!")
        return conn
    except Exception as e:
        print("An error occurred:", str(e))
        return None

# Function to extract quote numbers and names
def extract_quote_number_and_name(subject):
    #multiple_quote_pattern = r'(\d{6})(?:\s*/\s*(?=\(\s*Statements\s*\)|:))?'
    multiple_quote_pattern = r'(\d{6})(?:\s*/\s*(?=\d{6}))?'
    quote_numbers = re.findall(multiple_quote_pattern, subject)
    quote_number_and_name_pairs = []

    if len(quote_numbers) > 1:
        pattern = r'((?:\d{6}/)*\d{6})\s*:\s*(.*?)\s*\('
        matches = re.findall(pattern, subject)
        quote_numbers = []
        company_names = []

        for match in matches:
            quote_number = match[0].strip()
            company_name = match[1].strip() if match[1] else "unknown"
            if company_name == "unknown":
                # Find non-numerical text
                match = re.search(r'\b([A-Za-z\s]+)\(Statement\)', subject)
                if match:
                    # Use the matched text to replace "unknown"
                    company_name = match.group(1).strip().replace(' ', '_')
            
            # Insert to array
            quote_numbers.append(quote_number)
            company_names.append(company_name)

        print(quote_numbers)
        print(company_names)
        print(len(quote_numbers))
        print("Length of quote_numbers:", len(quote_numbers))
        print("Length of company_names:", len(company_names))

        for quote_number, company_name in zip(quote_numbers, company_names):
            quote_number_and_name_pairs.append((quote_number, company_name))

        print(quote_number_and_name_pairs)
        return quote_number_and_name_pairs

    elif len(quote_numbers) == 1:
        pattern = r'\b(\d{6})(?:\s*[:-])?\s*(.*)'
        regex = re.compile(pattern)
        match = regex.search(subject)

        if match:
            quote_number = match.group(1)
            company_name = match.group(2).strip().replace('(', '').replace(')', '').replace('/', '').replace(' ', '_').replace(':', '_') if match.group(2) else "unknown"
            return [(quote_number, company_name)]
        else:
            return []
    else:
        return []

# Function to process emails
def process_emails():
    mail = None
    try:
        EMAILS = gmail_setup()

        if mail:
            
            ## I need to normalise the emails stored and make sure i can perform these functions
            result_sender, data_sender = mail.search(None, '(FROM stratocal-client@outlook.com)')
            result_recipient, data_recipient = mail.search(None, '(TO stratocal-client@outlook.com)')
            result_cc, data_cc = mail.search(None, '(CC stratocal-client@outlook.com)')
            result_accounting, data_accounting = mail.search(None, '(FROM ap1_scl@stratocal.com)')

            email_ids_sender = data_sender[0].split() if data_sender else []
            email_ids_recipient = data_recipient[0].split() if data_recipient else []
            email_ids_cc = data_cc[0].split() if data_cc else []
            email_ids_accounting = data_accounting[0].split() if data_accounting else []

            email_ids = set(email_ids_sender + email_ids_recipient + email_ids_cc + email_ids_accounting)

            if email_ids:
                for latest_email_id in email_ids:
                    try:
                        result, data = mail.fetch(latest_email_id, '(RFC822)')
                        if result != 'OK':
                            print(f"Failed to fetch email ID {latest_email_id}")
                            continue

                        raw_email = data[0][1]
                        email_message = email.message_from_bytes(raw_email)
                        subject = email_message['Subject']
                        quote_numbers_and_names = extract_quote_number_and_name(subject)

                        if not quote_numbers_and_names:
                            print("Unable to extract quote number or name from email subject:", subject)
                            continue

                        save_folder = r'\\Server01\2tb\xampp\htdocs\stratocal\uploads'
                        if not os.path.exists(save_folder):
                            os.makedirs(save_folder)

                        received = f"{datetime.now().date()}"
                        dispatched = None
                        turnaround = None
                        sub_category = "Statements" if any(keyword in subject.lower() for keyword in ["statement", "statements"]) else "Correspondence"

                        conn = connect_to_mysql()
                        cursor = conn.cursor()

                        for quote_number, name in quote_numbers_and_names:
                            if not name:
                                print(f"Quote number {quote_number} has no associated name in subject: {subject}")
                                continue

                            email_file = download_email(mail, latest_email_id, save_folder, quote_number, name)
                            if email_file:
                                print("Email saved as:", email_file)

                                cursor.execute("INSERT INTO quotes (quoteno, name, received, dispatched, turnaround, sub_category, file_directory) VALUES (%s, %s, %s, %s, %s, %s, %s)",
                                                (quote_number, name, received, dispatched, turnaround, sub_category, email_file))
                                conn.commit()
                                print("Data inserted successfully for quote:", quote_number)
                            else:
                                print("Failed to save email.")

                        cursor.close()
                        conn.close()

                        mail.store(latest_email_id, '+FLAGS', '\\Deleted')
                    except Exception as e:
                        print("An error occurred while processing email:", str(e))
            else:
                print("No new emails to process.")
        else:
            print("Failed to connect to the email server.")
    except Exception as e:
        print("An error occurred:", str(e))

    return mail

# Main loop
while True:
    mail = process_emails()
    if mail:
        mail.expunge()
    time.sleep(120)
```
